```python
# imports
import pandas as pd
import numpy as np
import functools
from functools import reduce
```

```python
# read data
albumin = pd.read_csv('./cleaned_data/albumin.csv')
bp = pd.read_csv('./cleaned_data/bp.csv')
cholestrol = pd.read_csv('./cleaned_data/cholestrol.csv')
creatnine = pd.read_csv('./cleaned_data/creatnine.csv')
creatnine_clearance = pd.read_csv('./cleaned_data/creatnine_clearance.csv')
demo = pd.read_csv('./cleaned_data/demographics.csv')
```

```python
glycemia = pd.read_csv('./cleaned_data/glycemia.csv')
hdl = pd.read_csv('./cleaned_data/hdl.csv')
height = pd.read_csv('./cleaned_data/height.csv')
hemoglobin = pd.read_csv('./cleaned_data/hemoglobin.csv')
ldl = pd.read_csv('./cleaned_data/ldl.csv')
triglycerides = pd.read_csv('./cleaned_data/triglycerides.csv')
weight = pd.read_csv('./cleaned_data/weight.csv')
```

```python
diag = pd.read_csv('./cleaned_data/diag_cvd.csv')
```

```python
albumin = albumin.drop(columns=["Unnamed: 0", "microalbuminuria_date"])
bp = bp.drop(columns=["Unnamed: 0", "blood_pressure_date"])
cholestrol = cholestrol.drop(columns=["Unnamed: 0", "cholesterol_date"])
creatnine = creatnine.drop(columns=["Unnamed: 0", "creatinine_date"])
creatnine_clearance = creatnine_clearance.drop(columns=["Unnamed: 0", "creatinine_clearance_date"])
demo = demo.drop(columns=["Unnamed: 0", "IGNORE_THIS_COLUMN_1", "IGNORE_THIS_COLUMN_2", "DVP_REG_CODE", "DVP_PRO_CODEO", "TYPE_USER"])
diag = diag.drop(columns=["Unnamed: 0", "Diagnosis_Date"])
glycemia = glycemia.drop(columns=["Unnamed: 0", "glycemia_date"])
hdl = hdl.drop(columns=["Unnamed: 0", "HDL_high-density_lipoprotein_date"])
height = height.drop(columns=["Unnamed: 0", "height_date"])
hemoglobin = hemoglobin.drop(columns=["Unnamed: 0", "glycated_hemoglobin_date"])
ldl = ldl.drop(columns=["Unnamed: 0", "LDL_low-density_lipoprotein_date"])
triglycerides = triglycerides.drop(columns=["Unnamed: 0", "triglycerides_date"])
weight = weight.drop(columns=["Unnamed: 0", "weight_date"])
```

```python
albumin = albumin.drop_duplicates(subset=['PK_PATIENT'], keep='last')
bp =bp.drop_duplicates(subset=['PK_PATIENT'], keep='last')
cholestrol = cholestrol.drop_duplicates(subset=['PK_PATIENT'], keep='last')
creatnine = creatnine.drop_duplicates(subset=['PK_PATIENT'], keep='last')
creatnine_clearance = creatnine_clearance.drop_duplicates(subset=['PK_PATIENT'], keep='last')
demo = demo.drop_duplicates(subset=['PK_PATIENT'], keep='last')
diag = diag.drop_duplicates(subset=['PK_PATIENT'], keep='last')
glycemia = glycemia.drop_duplicates(subset=['PK_PATIENT'], keep='last')
hdl = hdl.drop_duplicates(subset=['PK_PATIENT'], keep='last')
height = height.drop_duplicates(subset=['PK_PATIENT'], keep='last')
hemoglobin = hemoglobin.drop_duplicates(subset=['PK_PATIENT'], keep='last')
ldl = ldl.drop_duplicates(subset=['PK_PATIENT'], keep='last')
triglycerides = triglycerides.drop_duplicates(subset=['PK_PATIENT'], keep='last')
weight = weight.drop_duplicates(subset=['PK_PATIENT'], keep='last')
```

```python
frames = [albumin, bp, cholestrol, creatnine, creatnine_clearance, demo, diag, glycemia, 
         hdl, height, hemoglobin, ldl, triglycerides, weight]

df_final = functools.reduce(lambda left,right: pd.merge(left,right,on='PK_PATIENT'), frames)
```
